```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = new["Summary"].values.astype("U")
```

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(documents)
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(10,30): 
    ke = KMeans(n_clusters=k,random_state=1)
    ke.fit(features)
    wcss.append(ke.inertia_)
```

```python
plt.plot(range(10,30),wcss, 'bx-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
ke = KMeans(n_clusters=12,random_state=1)
labels = ke.fit_predict(features)
```

```python
new["label"] = labels
new.head()
```
